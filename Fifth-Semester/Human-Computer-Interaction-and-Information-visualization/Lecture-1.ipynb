
## 🌍 What is HCI (Human-Computer Interaction)?

* It’s about **how humans interact with machines**.
* Three parts:

  1. **Humans** → with their work, abilities, and limits.
  2. **Machines** → with modes of operation, limits, and possibilities.
  3. **Interface** → where humans and machines meet.

👉 Example: pressing a button on your phone (input) → system responds on screen (output).

💡 Almost **half of software development** is about UI and interaction design.

---

## 🕰️ Historical Context

* **1940s**: Computers like Harvard Mark I – big, heavy, only used by specialists.
* **1970s onwards**: PCs and new interaction paradigms arrived → timesharing, networking, graphical displays.
* **Today**: Interaction includes handhelds, cloud, VR, robots, touchscreens.
* **Paradigms**:

  * 1950s → batch processing
  * 1960s → timesharing
  * 1970s → networking, email
  * 1980s → graphical interfaces
  * 1990s → WWW
  * Now → ubiquitous computing, VR, robot interaction

---

## 🎨 Design & Interaction

* Computers must be **intuitive, safe, and logical**.
* Users shouldn’t need to “think about how to use it.”
* Example: operating a car vs. a complex machine → design matters.
* Design sits **between technology and human purposes** (bridges the two).

---

## 🤖 Directions in HCI

* Human-robot interaction (e.g., androids like Geminoid in Japan).
* Emotional responses and personal relations with machines are becoming real (ethical debates here).

## 🔬 Multi-disciplinary Nature

HCI involves:

* **Computer Science** (software, systems).
* **Psychology** (cognition, learning, social behavior).
* **Design** (graphics, instructional).
* **Human factors/ergonomics**.
* **Anthropology, sociology, AI**.

---

## 📖 Definitions of HCI

1. **Practical**: processes and actions through which a human uses and interacts with a computer.
2. **Research focus**: design, evaluation, and implementation of interactive systems.

Key elements in definition:

* **Human** → single or groups of users.
* **Computer** → not just PCs, also embedded systems, large systems.
* **Interaction** → dialogue, feedback, tasks, mapping input/output.

---

## ✨ Key Concept 1: Usability

* Usability = how effective, efficient, and satisfying a system is.
* **Not** the same as “user-friendly” (too vague).
* Measured with **prototyping, evaluation, standards**.

---

## 📊 Information Visualization (InfoVis)

* **Goal**: make complex data understandable and explorable.
* **Visualization** = creating images/models to help us think about data.
* **InfoVis vs. SciVis**:

  * InfoVis → abstract data (text, stats, networks), designed visual space.
  * SciVis → physical/spatial data (3D models, medical scans).

---

## 🔑 Key Concepts (HCI + InfoVis)

* **Cognetics** (how people think and perceive).
* **Affordance** (what actions an object suggests, e.g., a button invites pressing).
* **Visibility** (clear options).
* **Task abstraction**.
* **Usability & evaluation**.
* **Prototyping**.
* **Data & transformations** (for InfoVis).

---

## ⚠️ Disaster Example

* **1988 Iran Air Flight 655**:

  * USS Vincennes shot down a plane due to misinterpreting radar displays.
  * Problem: poor **information visualization** (clutter, mapping issues).
  * Lesson: HCI/InfoVis can literally save lives.
