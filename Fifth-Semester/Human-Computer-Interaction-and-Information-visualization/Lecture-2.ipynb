## 🎯 Core Approach in Lecture 2

* **User-Centred Design**:

  * Interfaces should adapt to users, not the other way around.
  * Quote (Donald Norman): “Talking to users is not a luxury, it’s a necessity.”
* Goal: **optimize human + computer performance together as a system.**

---

## ⚡ Usability (in depth)

* **Poor interface design leads to**: mistakes, inaccessibility, frustration, low productivity, even system rejection.
* **ISO 9126 software quality metrics**: functionality, reliability, usability, efficiency, maintainability, portability.
* **Usability definition (Nielsen, 2003)**:

  * Ease of use + methods to improve ease-of-use.
  * Requires testing with *real users*.
* **Usability components**:

  1. **Learnability** (easy to learn, memorable).
  2. **Efficiency/Throughput** (speed, low error rate).
  3. **Flexibility** (customizable, suitable for skill level).
  4. **Satisfaction** (subjective feelings).

👉 Usability goals (objective) vs. User experience goals (subjective). Example: safety vs. fun.

## 🧠 Human Psychology in HCI

### Cognitive Psychology basics

* Humans process information via **perception → processing → action**.
* Input: senses (vision, hearing, touch, etc.).
* Processing: reasoning, problem-solving, memory.
* Output: actions + emotions.

### Cognition

* **Definition**: psychological processes to acquire, organize, and use knowledge.
* **Two modes**:

  * **Experiential (unconscious)** → fast, automatic, little effort.
  * **Reflective (conscious)** → slow, deliberate, decision-making.

### Left/Right Brain (simplified model)

* Left: words, logic, sequential tasks.
* Right: images, spatial, holistic patterns.
  ⚠️ Modern neuroscience says “right-brain vs left-brain people” is a myth — both hemispheres work together.

## 🗂️ Memory Models

### Types of Memory

* **Sensory memory**: 0.2–2 seconds (raw input buffers: vision, sound, touch).
* **Short-term / Working memory (WM)**:

  * 20–30 sec with rehearsal.
  * Limited capacity → “7±2 chunks” (Miller, 1956) → now thought to be 3–4 items.
* **Long-term memory (LTM)**:

  * Unlimited, stores semantic (facts, concepts) + episodic (personal events).
  * Access depends on refresh, associations, context.
  * Represented as semantic networks.

### Model Human Processor (Card, Moran & Newell, 1983)

* **3 subsystems**: perceptual, cognitive, motor.
* Each has cycle times (perception \~100ms, cognition \~70ms, motor \~70ms).
* Implication: human capacity is limited → designers must simplify, chunk, reduce overload.

## 🧩 Psychological Factors

* **Closure** → people want to finish tasks to free working memory.

  * Example: ATM or online payment must be short + clear.
* **Attitude & Anxiety** → negative feelings reduce memory capacity and slow learning.
* **Control** → beginners want guidance; experts want speed + control.
* **Recognition vs Recall** → easier to recognize than recall. Use menus, icons, chunking.

---

## 👀 Attention

* **Definition**: focusing on one thing while ignoring others.
* **Types**: orienting, expecting, searching, filtering.
* **Cocktail party effect**: filtering background noise, but reacting when hearing your name.
* **Locus of attention**: humans can only have **one main focus at a time**.
* **Implication for UI**:

  * Structure information (chunking, grouping).
  * Use spatial/temporal cues.
  * Highlight with color, flashing, sounds, or icons.

---

## 🧠 Cognitive Processing & Meaningfulness

* **Automated processing** (fast, unconscious).
* **Controlled processing** (slow, conscious).
* **Meaningful information** is remembered better.

  * Familiar, imageable, contextual words help (e.g., “Sleep” vs. “Evaluate”).
* Other influences: **culture (color meanings), context, emotional state**.

---

## 😊 Emotion in Interaction

* **Theories**:

  * James-Lange: emotion = interpretation of body response.
  * Cannon: emotion = direct psych response.
  * Schachter-Singer: emotion = body response + situation context.
* **Affect**: biological response that influences performance.

  * Positive → more creativity, easier problem-solving.
  * Negative → slower, narrower thinking.
* **Design implication**:

  * Stress makes tasks harder.
  * Aesthetically pleasing, rewarding UIs increase positive affect → more tolerance.
