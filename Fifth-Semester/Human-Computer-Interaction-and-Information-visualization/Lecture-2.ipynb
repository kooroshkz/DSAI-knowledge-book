## âš¡ Usability (in depth)

* **Poor interface design leads to**: mistakes, inaccessibility, frustration, low productivity, even system rejection.
* **ISO 9126 software quality metrics**: functionality, reliability, usability, efficiency, maintainability, portability.
* **Usability definition (Nielsen, 2003)**:

* **Usability components**:

  1. **Learnability** (easy to learn, memorable).
  2. **Efficiency/Throughput** (speed, low error rate).
  3. **Flexibility** (customizable, suitable for skill level).
  4. **Satisfaction** (subjective feelings).

## ðŸ‘€ Attention

* **Definition**: focusing on one thing while ignoring others.
* **Types**: orienting, expecting, searching, filtering.
* **Cocktail party effect**: filtering background noise, but reacting when hearing your name.
* **Locus of attention**: humans can only have **one main focus at a time**.
* **Implication for UI**:

  * Structure information (chunking, grouping).
  * Use spatial/temporal cues.
  * Highlight with color, flashing, sounds, or icons.
